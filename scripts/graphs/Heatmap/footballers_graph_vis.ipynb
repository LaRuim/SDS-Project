```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt

data = pd.read_csv("players_20.csv", encoding='cp1252')
df = pd.DataFrame(data)
bmi = []
for ind in df.index:
    bmi.append((10000*df['weight_kg'][ind])/(df['height_cm'][ind]*df['height_cm'][ind]))
df['bmi'] = bmi
plt.boxplot(df.bmi)
plt.show()
# plt.boxplot(data.height_cm)
# plt.boxplot(data.weight_kg)
```
